# CLIENT

```python
import numpy as np
import tenseal as ts
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
class Client:
    
    def __init__(self, user_id):     
        
        #encryption parameters
        # encryption parameters
        poly_mod_degree = 2 ** 14
        power = 30
        coeff_mod_bit_sizes = [40] + ([power] * 6) + [40]

        # create TenSEALContext
        ctx_eval = ts.Context(ts.SCHEME_TYPE.CKKS,poly_mod_degree,-1,coeff_mod_bit_sizes=coeff_mod_bit_sizes,)

        # scale of ciphertext to use
        ctx_eval.global_scale = 2 ** power

        # this key is needed for doing dot-product operations
        ctx_eval.generate_galois_keys()
        
        
        self.user_id = user_id   
        self.client_context = ctx_eval
        self.public_context = ctx_eval.copy()
        self.sample = None
        self.encrypted_result = None
        self.enc_sample = None
        
                    
    """
    Raw sample data:
    input_data = np.array(shape = (1, 30), dtype='f')
    """
    def receive_sample(self, input_data):
        self.sample = input_data
    
    def receive_encrypted_result(self, data):
        self.encrypted_result = data
        
   

    def encrypt_sample(self, sample):
        assert sample is not None, "Sample can not be None!"
        self.enc_sample = ts.ckks_vector(self.public_context, self.sample)
        self.sample = None
        self.public_context.make_context_public()
    
    
    def decrypt_sample(self, encrypted_output):
        return encrypted_output.decrypt(secret_key=self.client_context.secret_key())
    
    def client_query(self):
        return {
            "data": self.enc_sample.serialize(),
            "context": self.public_context.serialize(),
        }
    
```

# SERVER

```python
class Server():
    def __init__(self):
        self.dense1_kernel = np.load("./ckks_data/dense1_kernel.npy")
        self.dense1_bias = np.load("./ckks_data/dense1_bias.npy")
        self.dense2_kernel = np.load("./ckks_data/dense2_kernel.npy")
        self.dense2_bias = np.load("./ckks_data/dense2_bias.npy")
        self.dense3_kernel = np.load("./ckks_data/dense3_kernel.npy")
        self.dense3_bias = np.load("./ckks_data/dense3_bias.npy")
    
        self.enc_sample = None
        
    def forward(self, enc_x: ts.CKKSVector) -> ts.CKKSVector:
        #dense1
        out = enc_x.mm_(self.dense1_kernel) + self.dense1_bias        
        #square
        out = out.square_()        
        #dense2
        out = out.mm_(self.dense2_kernel) + self.dense2_bias        
        #square
        out = out.square_()        
        #dense3
        out = out.mm_(self.dense3_kernel) + self.dense3_bias
        
        return out
    
    def get_query(self, query) -> ts.CKKSVector:
        try:
            public_context = ts.context_from(query["context"])
            enc_sample = ts.ckks_vector_from(public_context, query["data"])
        except:
            raise DeserializationError("cannot deserialize context or ckks_vector")
        try:
            _ = public_context.galois_keys()
        except:
            raise InvalidContext("the context doesn't hold galois keys")
        
        return enc_sample
    
```

```python
server = Server()
enc_sample = server.get_query(query)
```
